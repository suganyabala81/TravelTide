```python
# Create query
query = '''
--Travel frequency,anddays for married users with children
select
u.user_id,
u.home_country,
count(f.trip_id)as total_flights,
 SUM(EXTRACT(DAY FROM (f.return_time - f.departure_time))) AS travel_days
   from users u
   join sessions s
   on u.user_id = s.user_id
   join flights f
   on s.trip_id = f.trip_id
 where u.married = True And u.has_children = true
  group by u.user_id,u.home_country
   order by total_flights desc;
'''

pd.read_sql(query, con=connection)
```
